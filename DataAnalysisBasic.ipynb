**Data Normalization**\

1)Simple feature scaling:**new=old/max**(0->1)\

3)Z-score:**new=(old-avg)/s.d.**(center point is 0 and most at -3 -> +3)(df.mean()->avg,df.std->s.d.)

**Correlation**\

**Pearson correlation**\
Correlation coefficient:+1 Large positive correlation\
                        -1 Large negative correlation\
                         0 No relationship\
P-value:<0.001 Strong,<0.05 Moderate,<0.1 weak,>0.1 no certainty\
